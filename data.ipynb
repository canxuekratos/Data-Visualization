```python
import numpy as np
import pandas as pd
#画图包
from pyecharts.charts import Map,Bar,Line,Pie,Page,Timeline,Grid
from pyecharts.globals import ThemeType
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
#在jupyter notebook演示
from pyecharts.globals import CurrentConfig, NotebookType
# 配置对应的环境类型
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
CurrentConfig.ONLINE_HOST='https://assets.pyecharts.org/assets/'
 
```

```python
df = pd.read_excel(
    '人口自然增长率.xls',
    index_col=0,
    header=3)
df.dropna(how='any', inplace=True)
df
```

```python
df2=df.T.sort_values('2021年',ascending=True)
df2
```

```python
def _map(year:str):
    # 处理数据
    data = [[name,[value,name]] for name,value in zip(df2.index.values,df2[f'{year}'])]
    # 地图
    chart_map = (
        Map()
        .add(
            series_name='',
            data_pair=data,
            is_map_symbol_show=False, # 显示图形标记，即小红点
            label_opts=opts.LabelOpts(is_show=False), # 不显示地名
            itemstyle_opts={  # 图元样式配置项
            "normal": {
                # 显示的样式
                "areaColor": "#CED8F6",
                "borderColor": "#404a59"
            },
            "emphasis": {
                # 鼠标选中后显示的样式
                "lable": {"show": Timeline},
                "areaColor": "rgba(255,255,255, 0.5)"
            },
            }
        )
        .set_global_opts(
                title_opts=opts.TitleOpts(
                    title=f'{year}各省人口自然增长率',
                    pos_left="center",
                    pos_top="top",
                    title_textstyle_opts=opts.TextStyleOpts(
                        font_size='35',
                        color="rgba(155,155,155,1)"
                    )
                ),
            )
    )
    return  chart_map
```

```python
def _line():
    x1 = [i for i in df2.keys()][::-1]
    y1 = [i for i in df2.loc['上海市']][::-1]
    y2 = [i for i in df2.loc['北京市']][::-1]
    y3 = [i for i in df2.loc['天津市']][::-1]
    line_chart=(
        Line()
        .add_xaxis(xaxis_data=x1)
        .add_yaxis(
            series_name="上海市",
            y_axis=y1,
            is_connect_nones=True
        )
        .add_yaxis(
            series_name="北京市",
            y_axis=y2,
            is_connect_nones=True,
            symbol="arrow",
        )
        .add_yaxis(
            series_name="天津市",
            y_axis=y3,
            is_connect_nones=True,
            symbol="rect",
        )
        .set_global_opts(title_opts=opts.TitleOpts(title="2002-2021人口增长率"),
                        yaxis_opts=opts.AxisOpts(
                            name='人口增长率',
                            axislabel_opts=opts.LabelOpts(formatter="{value}%")))
    )
    return line_chart
```

```python
def _bar(year:str):
    # 处理数据
    df2=df.T.sort_values(year,ascending=True)
    bar_data = [[name,[value,name]] for name,value in zip(df2.index.values,df2[f'{year}'])]
    lst = [value for value in df2[f'{year}']]
    ave=sum(lst)/len(lst)
    num=len([item[1][0] for item in bar_data if item[1][0]>0])*17
    num2=600-num
    # 柱状图
    bar1_chart = (
        Bar()
            .add_xaxis(
                xaxis_data=[item[0] for item in bar_data if item[1][0]>=0] # x轴数据, 格式为[1,2,3,4,5,]
            )
            .add_yaxis(
                series_name='',
                y_axis=[item[1][0] for item in bar_data if item[1][0]>=0],   # y轴数据, 格式为[1,2,3,4,5,]
                
                bar_width=11,
                label_opts=opts.LabelOpts(
                    #font_size=18,
                    position='right',
                    formatter="{b}: {c}",   # 显示样式
                )   
            )
            
            .set_global_opts(
                # 不显示坐标轴的 标签
                xaxis_opts=opts.AxisOpts(
                    min_=0,
                    max_=10,
                    is_inverse=False,
                    is_show=False,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False)),
                yaxis_opts=opts.AxisOpts(
                    is_show=False,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False)),
                legend_opts=opts.LegendOpts(is_show=True,item_gap=5),
                tooltip_opts=opts.TooltipOpts(is_show=False),
                visualmap_opts=opts.VisualMapOpts(
                    orient='vertical',
                    is_calculable=True,
                    dimension=0,
                    pos_right='10%',  # 组件位置
                    pos_bottom='bottom',  # 组件位置
                    range_text=['max', 'min'],  # 组件两端的文本
                    range_color=["lightskyblue", "yellow", "orangered"],  # 组件过渡颜色
                    min_=-1,  # 组件最小值
                    max_=12,  # 组件最大值
                    textstyle_opts=opts.TextStyleOpts(
                        color="rgba(0,0,0,0.5)",  # 文字颜色
                        font_size=15
                    ),
                )
            )
            .reversal_axis()
    )
    bar2_chart = (
        Bar()
            .add_xaxis(
                xaxis_data=[item[0] for item in bar_data if item[1][0]<0] # x轴数据, 格式为[1,2,3,4,5,]
            )
            .add_yaxis(
                series_name='',
                y_axis=[item[1][0] for item in bar_data if item[1][0]<0],   # y轴数据, 格式为[1,2,3,4,5,]
                bar_width=11,
                category_gap=0,
                label_opts=opts.LabelOpts(
                    #font_size=18,
                    position='right',
                    formatter="{b}: {c}",   # 显示样式
                )
                
            )
            .set_global_opts(
                # 不显示坐标轴的 标签
                xaxis_opts=opts.AxisOpts(
                    min_=-10,
                    max_=0,
                    is_inverse=True,
                    is_show=False,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False)),
                yaxis_opts=opts.AxisOpts(
                    is_show=False,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False)),
                tooltip_opts=opts.TooltipOpts(is_show=False),
                graphic_opts=[
                opts.GraphicGroup(
                    graphic_item=opts.GraphicItem(
                        left=300,
                        bottom=60,
                    ),
                    children=[
                        opts.GraphicText(
                            graphic_item=opts.GraphicItem(
                                 # 设置字形的 大小，位置
                                left="center",
                                top="center",
                            ),
                            graphic_textstyle_opts=opts.GraphicTextStyleOpts(
                                # 设置水印、字体等，见（2）
                                text=f"平均值：{ave:.2f}",
                                font="bold 18px Microsoft YaHei",
                                graphic_basicstyle_opts=opts.GraphicBasicStyleOpts(
                                    # 设置字体颜色
                                    fill="rgb(0,0,0)"
                                ),
                            ),
                        ),
                    ],
                )
            ],
            #
        )
            .reversal_axis()
        
    )
    bar_chart = (
        Grid(init_opts=opts.InitOpts(height='600px'))
            .add(bar2_chart, grid_opts=opts.GridOpts(pos_left="10", pos_right="45%",pos_top=f'{num}px'))
            .add(bar1_chart, grid_opts=opts.GridOpts(pos_left="10", pos_right="45%",pos_bottom=f'{num2}px'))
            .add(_map(year), grid_opts=opts.GridOpts())
    )
    return bar_chart
```

```python
_bar('2002年').render_notebook()
```

```python
def timeline():
    # 获取日期 排序
    #time_list=df2.keys().tolist()
    #time_list=time_list[::-1]
    time_list2=['2002年','2003年','2004年','2005年','2006年','2007年','2008年','2009年','2010年',
                '2011年','2012年','2013年','2014年','2015年','2016年','2017年','2018年','2019年','2020年']
    timeline = Timeline(init_opts=opts.InitOpts(height='600px'))
    for date in time_list2:
        #charts = _map(date)  # 获取图表
        #charts = _map(date)  # 获取图表
        bar_charts = _bar(date)  # 获取图表
        #timeline.add(charts, time_point=str(date))
        timeline.add(bar_charts, time_point=str(date))
    timeline.add_schema(
        is_auto_play=True,
        is_timeline_show=True
    )
    return timeline
```

```python
_timeline = timeline()
_timeline.render_notebook()
```

```python
df3 = pd.read_excel(
    '人口数据.xls',
    index_col=0,
    header=2)
df3.dropna(how='any', inplace=True)
df4=df3.T
df4
```

```python
def _line2():
    x1 = [i for i in df4.keys()][::-1]
    y1 = [i for i in df4.loc['人口出生率(‰)']][::-1]
    y2 = [i for i in df4.loc['人口死亡率(‰)']][::-1]
    y3 = [i for i in df4.loc['人口自然增长率(‰)']][::-1]
    line_chart=(
        Line()
        .add_xaxis(xaxis_data=x1)
        .add_yaxis(
            series_name="人口出生率",
            y_axis=y1,
            is_connect_nones=True,
            is_smooth=True,
        )
        .add_yaxis(
            series_name="人口死亡率",
            y_axis=y2,
            is_connect_nones=True,
            symbol="arrow",
            is_smooth=True,
        )
        .add_yaxis(
            series_name="人口自然增长率",
            y_axis=y3,
            is_connect_nones=True,
            symbol="rect",
            is_smooth=True,
        )
        .set_global_opts(
                title_opts=opts.TitleOpts(
                    title="2002-2021全国人口出生率/死亡率/自然增长率",
                    pos_left="center",
                    pos_top="bottom",
                ),
                yaxis_opts=opts.AxisOpts(
                    name='',
                    axislabel_opts=opts.LabelOpts(formatter="{value}(‰)")))
    )
    return line_chart
_line2().render_notebook()
```

```python
page=Page(layout=Page.DraggablePageLayout,page_title='人口增长率')
page.add(
    timeline(),
    _line(),
    _line2())
page.render('page.html')
```
