```python
import numpy as np
import pandas as pd
#画图包
from pyecharts.charts import Map,Bar,Line,Pie,Page,Timeline,Grid
from pyecharts.globals import ThemeType
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
#在jupyter notebook演示
from pyecharts.globals import CurrentConfig, NotebookType
# 配置对应的环境类型
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
CurrentConfig.ONLINE_HOST='https://assets.pyecharts.org/assets/'
```

```python
df = pd.read_excel(
    '人口数据.xls',
    index_col=0,
    header=2)
df.dropna(how='any', inplace=True)
df
```

```python
df2=df.T.sort_values('2021年',ascending=True)
df2
```

```python
x=[i for i in df2.keys()][::-1][1:]
y1=[i for i in df2.loc['上海市']][::-1][1:]
y2=[i for i in df2.loc['上海市']][1:][::-1]
list(map(lambda x:(x[0]-x[1])/x[1]*100,zip(y1,y2)))
```

```python
def _line():
    x=[i for i in df2.keys()][::-1][1:]
    y1=[i for i in df2.loc['上海市']][::-1][1:]
    y2=[i for i in df2.loc['上海市']][1:][::-1]
    y=list(map(lambda x:(x[0]-x[1])/x[1]*100,zip(y1,y2)))
    line_chart=(
        Line()
        .add_xaxis(xaxis_data=x)
        .add_yaxis(
            series_name="上海市",
            y_axis=y,
            is_connect_nones=True
        )
        .set_global_opts(title_opts=opts.TitleOpts(title="2003-2021经济"),
                        yaxis_opts=opts.AxisOpts(
                            name='增长率',
                            axislabel_opts=opts.LabelOpts(formatter="{value}%")))
    )
    return line_chart
```

```python
_line().render_notebook()
```
