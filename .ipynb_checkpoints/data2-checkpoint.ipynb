```python
import numpy as np
import pandas as pd
#画图包
from pyecharts.charts import Map,Bar,Line,Pie,Page,Timeline,Grid
from pyecharts.globals import ThemeType
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
```

```python
df = pd.read_excel(
    '人口数据.xls',
    index_col=0,
    header=2)
df.dropna(how='any', inplace=True)
df2=df.T
df2
```

```python
def _line():
    x1 = [i for i in df2.keys()][::-1]
    y1 = [i for i in df2.loc['人口出生率(‰)']][::-1]
    y2 = [i for i in df2.loc['人口死亡率(‰)']][::-1]
    y3 = [i for i in df2.loc['人口自然增长率(‰)']][::-1]
    line_chart=(
        Line()
        .add_xaxis(xaxis_data=x1)
        .add_yaxis(
            series_name="人口出生率",
            y_axis=y1,
            is_connect_nones=True,
            is_smooth=True,
        )
        .add_yaxis(
            series_name="人口死亡率",
            y_axis=y2,
            is_connect_nones=True,
            symbol="arrow",
            is_smooth=True,
        )
        .add_yaxis(
            series_name="人口自然增长率",
            y_axis=y3,
            is_connect_nones=True,
            symbol="rect",
            is_smooth=True,
        )
        .set_global_opts(
                title_opts=opts.TitleOpts(
                    title="2002-2021全国人口出生率/死亡率/自然增长率",
                    pos_left="center",
                    pos_top="bottom",
                ),
                yaxis_opts=opts.AxisOpts(
                    name='',
                    axislabel_opts=opts.LabelOpts(formatter="{value}(‰)")))
    )
    return line_chart
```
