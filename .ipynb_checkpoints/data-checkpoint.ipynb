```python
import numpy as np
import pandas as pd
#画图包
from pyecharts.charts import Map,Bar,Line,Pie,Page,Timeline
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
#在jupyter notebook演示
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.NTERACT
```

```python
df = pd.read_excel(
    '人口自然增长率.xls',
    index_col=0,
    header=3)
df.dropna(how='any', inplace=True)
df
```

```python
df2=df.T.sort_values('2021年',ascending=True)
df2
```

```python
[[name,[value,name]] for name,value in zip(df2.index.values,df2[f'{2021}年'])]
```

```python
def _map(year):
    # 处理数据
    data = [[name,[value,name]] for name,value in zip(df2.index.values,df2[f'{year}年'])]
    # 地图
    chart_map = (
        Map()
        .add(
            series_name='',
            data_pair=data,
            is_map_symbol_show=False, # 显示图形标记，即小红点
            label_opts=opts.LabelOpts(is_show=False) # 不显示地名
        )
    )
    return  chart_map
```

```python
chart_map = _map(2021)
chart_map.render_notebook()
```
